```python
import pandas as pd
import re
import pandas_profiling
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_file = 'Data/result.csv'
data = pd.read_csv(data_file)
data
```

```python
pandas_profiling.ProfileReport(data)
```

```python
def preprocessor(text):
    """ Return a cleaned version of text
    """
    # Remove HTML markup
    text = re.sub('<[^>]*>', '', text)
    # Save emoticons for later appending
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)', text)
    # Remove any non-word character and append the emoticons,
    # removing the nose character for standarization. Convert to lower case
    text = (re.sub('[\W]+', ' ', text.lower()) + ' ' + ' '.join(emoticons).replace('-', ''))
    
    return text
def preprocessor_genre(text):
    text = text.strip()
    text = text.replace(' ','')
    text = text.lower()
    return text
```

```python
dataset = pd.DataFrame()
```

```python
new_columns = ['Titles', 'TitlesVN', 'Release', 'Certificate', 'Runtime', 'Genre',
       'Rates', 'Directors']
```

```python
for index, column in enumerate(data.columns):
    if column in ['rates','certificate'] :
        dataset[new_columns[index]] = data[column]
    elif column == 'genre':
        dataset[new_columns[index]] = data[column].apply(preprocessor_genre)
    else:
        dataset[new_columns[index]] = data[column].apply(preprocessor)
```

```python
# to seperate genre column in dataset
seperated_genres = dataset['Genre'].str.get_dummies(',')

# to concatenate two dataframes
dataset = pd.concat([dataset, seperated_genres], axis = 1, sort = False)
```

```python
dataset.head()
```

```python
list_genre = dataset['Genre'].str.split(',',expand = True)
list_genre = list_genre.applymap(str)
```

```python
# extracting Genres from the df
genres = []
for i in range(0,3):
    list_genre[i]
    for j in range(0, 100) :
        list_genre[i][j] = list_genre[i][j].replace(' ','')
        if (list_genre[i][j] not in genres) and list_genre[i][j] != 'None' and list_genre[i][j] != 'nan':
            genres.append(list_genre[i][j])
        else:
            pass  
genres
```

```python
genre_mv_sum = []
for column in genres:
    num = dataset[column][dataset[column] == 1].sum()
    genre_mv_sum.append(num)
```

```python
data_genres = pd.DataFrame({
    'Genres' : genres,
    'Toltal' : genre_mv_sum}
)
```

```python
plt.figure(figsize = (20, 10))
sns.barplot(data = data_genres,
           y = data_genres['Genres'],
           x = data_genres['Toltal']
)
```

```python
rake = Rake()
```

```python
dataset['Directors'] = dataset['Directors'].fillna(' ')
dataset['Directors'] = dataset['Directors'].astype(str)
# dataset['cast'] = dataset['cast'].fillna(' ')
# dataset['cast'] = dataset['cast'].astype(str)
```

```python
dataset.head()
```

```python
key_notes = []
for index,row in dataset.iterrows():
    director =  row['Directors']
    genre = " ".join(row['Genre'].split(',')).lower()
    plot = row['Titles'] 
    rake.extract_keywords_from_text(plot)
    keyword_score = rake.get_word_degrees()
    keyword_score = ' '.join(list(keyword_score.keys()))
    key_note = genre + ' ' + ' ' + director + ' ' + plot
    key_notes.append(key_note)

    
dataset['key_notes'] = key_notes
recommend = dataset[['Titles','key_notes']]
recommend.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(dataset['key_notes'])
```

```python
def recommend_movie(query):
    Y_tfidf = vectorizer.transform([query])
    sims = cosine_similarity(Y_tfidf, X_tfidf)
    movies = []
    for i in range(5):
        idx_sims_max = np.argmax(sims)
        movies.append(recommend['Titles'][idx_sims_max])
        sims[0][idx_sims_max] = -1
    return movies
```

```python
def rec():
    try:
        i = 1
        while(i > 0):
            query = input("Enter The Name of a Movie or Tv Show: ")
            if query.lower() == 'quit':
                break
            else:
                print(recommend_movie(query))

    except KeyboardInterrupt:
        print("The movie or Tv Show does not exist\n")
        rec()

    except IndexError:
        print("The movie or Tv Show does not exist\n")
        rec()
        

print("To exit Enter \"quit\" \n")
rec()
```
